```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df=pd.read_csv(r'./company-size.csv')

data=df
```

```python
sns.scatterplot(data=data,x="marketcap",y="anrevenue",size="size",legend=False)
plt.show()
```

```python
plt.rcParams['figure.figsize']=[8,8]
sns.scatterplot(data=data,x="marketcap",y="anrevenue",size="size",legend=False)
plt.show()
```

```python
sns.scatterplot(data=data,x="marketcap",y="anrevenue",size="size")
plt.show()
```

```python
sns.scatterplot(data=data,x="marketcap",y="anrevenue",size="size",hue="sector")
plt.show()
```

```python
sns.scatterplot(data=data,x="marketcap",y="anrevenue",size="size",hue="sector", alpha=0.7)
plt.show()
```
