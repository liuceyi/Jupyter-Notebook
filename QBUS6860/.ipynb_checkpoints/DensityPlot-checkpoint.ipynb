```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="darkgrid")
df=sns.load_dataset('iris')
```

```python
sns.kdeplot(df['sepal_width'])
plt.show()
```

```python
sns.kdeplot(df['sepal_width'],shade=True)
plt.show()
```

```python
fig=sns.kdeplot(df['sepal_width'],shade=True,color="b")
fig=sns.kdeplot(df['sepal_length'],shade=True,color="r")
plt.show()
```

```python
sns.histplot(data=df,x="sepal_length",color="red",label="Sepal Length", kde=True)
sns.histplot(data=df,x="sepal_width",color="blue",label="Sepal Width", kde=True)

plt.legend()
plt.show()
```

```python
fig,axs=plt.subplots(1,2,figsize=(8,5))

sns.histplot(data=df,x="sepal_width",kde=True,color="blue",ax=axs[0])
sns.histplot(data=df,x="sepal_length",kde=True,color="red",ax=axs[1])

plt.show()
```

```python
import pandas as pd
df=pd.read_csv(r'./company-stats.csv')
```

```python
sns.kdeplot(df['marketcap'])
plt.show()
```

```python
sns.kdeplot(df['marketcap'],shade=True)
plt.show()
```

```python
fig=sns.kdeplot(df['marketcap'],shade=True,color="b")
fig=sns.kdeplot(df['anrevenue'],shade=True,color="r")
plt.show()
```

```python
sns.histplot(data=df,x="anrevenue",color="red",label="Annual Revenue", kde=True)
sns.histplot(data=df,x="marketcap",color="blue",label="Market Cap", kde=True)

plt.legend()
plt.show()
```

```python
fig,axs=plt.subplots(1,2,figsize=(8,5))

sns.histplot(data=df,x="marketcap",kde=True,color="blue",ax=axs[1])
sns.histplot(data=df,x="anrevenue",kde=True,color="red",ax=axs[0])

plt.show()
```
