```python
data_results = data_results_raw.copy()
data_results.replace({r'\N':np.nan}, inplace=True)
data_results.sort_values('rank')
```

```python
data_temp = pd.merge(data_results, data_race, how='inner', on='raceId').dropna(subset=['fastestLapSpeed'])
data_temp['fastestLapSpeed'] = data_temp['fastestLapSpeed'].astype('float64')
data_temp[['year', 'fastestLapSpeed']]
```

```python
fig = go.Figure()
def rgbGradient(start_RGB, end_RGB, rate):
    r = start_RGB[0] + (end_RGB[0] - start_RGB[0]) * rate
    g = start_RGB[1] + (end_RGB[1] - start_RGB[1])* rate
    b = start_RGB[2] + (end_RGB[2] - start_RGB[2])* rate
    return 'rgb({r},{g},{b})'.format(r=r, g=g, b=b)

fastestLapSpeed_mean = data_temp.groupby('year')['fastestLapSpeed'].mean()
fastestLapSpeed_rate = (fastestLapSpeed_mean - fastestLapSpeed_mean.min()) / (fastestLapSpeed_mean.max() - fastestLapSpeed_mean.min())
for year in np.sort(data_temp['year'].unique()):
    fig.add_trace(go.Box(
        y=data_temp[data_temp['year'] == year]['fastestLapSpeed'],
        name=year.astype(str),
        marker_color=rgbGradient((255,182,193), (255,20,147), fastestLapSpeed_rate[year])
    ))
    
from sklearn.linear_model import LinearRegression
X = data_temp['year']
y = data_temp['fastestLapSpeed']
reg = LinearRegression().fit(X.to_numpy().reshape(-1, 1), y)
y_pred = reg.predict(X.to_numpy().reshape(-1, 1))
fig.add_trace(go.Scatter(
    y=y_pred,
    x=fastestLapSpeed_mean.index,
    mode='lines'
))
fig.update_layout(title_text="Box Plot of Fastest Lap Speed each year")

fig.show()
```
