```python
#import libraries

import numpy as np
import pandas as pd

from bokeh.layouts import column
from bokeh.models import ColumnDataSource, RangeTool
from bokeh.plotting import figure, show

#import data

data = pd.read_csv(r'Data/bitcoin.csv')
data.head()
```

```python
dates = np.array(data['date'], dtype=np.datetime64)
```

```python
price_diff = []
for i in data.index:
    if i == 0:
        price_diff.append(0)
        continue
    price_now = data.iloc[i, 1]
    price_lastday = data.iloc[i-1, 1]
    price_diff.append(price_now - price_lastday)

data['price_diff'] = price_diff
data
```

```python
source = ColumnDataSource(data = dict(date = dates, close = data['price'], diff = data['price_diff']))
```

```python
p = figure(height=300, width=800, tools="xpan", toolbar_location=None,
           x_axis_type="datetime", x_axis_location="above",
           background_fill_color="#FFF0F5", x_range=(dates[1], dates[10]))

# p.line('date', 'close', source=source)
p.line('date', 'price_diff', source = source)
p.yaxis.axis_label = 'Price'
p.xaxis.axis_label = 'Date'

select = figure(title="Drag the middle and edges of the selection box to change the range above",
                height=130, width=800, y_range=p.y_range,
                x_axis_type="datetime", y_axis_type=None,
                tools="", toolbar_location=None, background_fill_color="#FFF0F5")

range_tool = RangeTool(x_range=p.x_range)
range_tool.overlay.fill_color = "#efefef"
range_tool.overlay.fill_alpha = 0.2

# select.line('date', 'close', source=source)
select.line('date', 'price_diff', source=source)
select.ygrid.grid_line_color = None
select.add_tools(range_tool)
select.toolbar.active_multi = range_tool

show(column(p, select))
```
