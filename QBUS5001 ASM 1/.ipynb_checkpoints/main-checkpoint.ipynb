```python
OLS_reg = sm.OLS(y, sm.add_constant(X, prepend=True)).fit()
OLS_reg.summary()
```

```python
y_pred = OLS_reg.fittedvalues
y_true = data_std['PE_TTM']
plt.scatter(X, y_true, s = 5, label='Data')
plt.plot(X, y_pred, lw = 1.5, c = 'red',label = 'Model')
plt.xlabel('Net Profit Margin')
plt.ylabel('Std PE Ratio')
plt.legend(loc = 'best')
plt.title('Standardized Model')
plt.show()
```
