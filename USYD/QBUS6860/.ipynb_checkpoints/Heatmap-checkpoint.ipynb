```python
import seaborn as sns
import pandas as pd
import numpy as np

df=pd.DataFrame(np.random.random((8,8)),columns=["a","b","c","d","e","f","g","h"])
```

```python
p1=sns.heatmap(df)
```

```python
sns.heatmap(df,annot=True,annot_kws={"size":7})
```

```python
sns.heatmap(df,cmap="YlGnBu")
# plt.show()
```

```python
import matplotlib.pyplot as plt
sns.heatmap(df,cmap="YlGnBu")
plt.show()
```

```python
sns.heatmap(df,cmap="Blues")
plt.show()
```

```python
sns.heatmap(df,cmap="Greens")
plt.show()
```
