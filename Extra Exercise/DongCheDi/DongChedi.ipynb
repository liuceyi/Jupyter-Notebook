```python
import json
import requests
import pandas as pd
```

```python
url = r"https://www.dongchedi.com/motor/car_page/v4/get_entity_json/?car_id_list={}&city_name=%E6%B7%B1%E5%9C%B3&version_code=444".format(38236)
req = requests.get(url=url).text
req
```

```python
data = json.loads(req)
car_info = data['data']['car_info'][0]
car_info_pro = car_info['info']
properties = data['data']['properties']
data_list = []
for pro in properties:
  key = pro['key']
  text = pro['text']
  if key != '' and text != '' and key in car_info_pro.keys():
    data_list.append({'name':text, 'value':car_info_pro[key]['value']})

pd.DataFrame(data=data_list)
```
